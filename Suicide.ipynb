# 1. Demographic Analysis

```python
df.columns
```

```python
# Is there a significant difference in suicide rates between males and females? 
gender_deff=df.groupby('sex')['suicides/100k pop'].mean().reset_index()
gender_deff
```

```python
px.pie(gender_deff,names='sex',values='suicides/100k pop')
```

- Finding: There is a significant gender gap in suicide rates. On average, the suicide rate for males (approx. 20.24 per 100k) is nearly four times higher than that for females (approx. 5.39 per 100k). This suggests that males are a much higher-risk group globally.

```python
comparison = df.groupby(['sex', 'new_age','age'])['suicides/100k pop'].mean().reset_index()
comparison
```

```python
px.line(comparison, x='age', y='suicides/100k pop', color='sex',title='Suicide Rates by Age Group')
```

- Finding: Suicide rates show a strong positive correlation with age. Both males and females experience an increase in suicide rates as they get older, with the 75+ years age group consistently recording the highest rates.

```python
# Has this gap closed or widened over the years?

gap_trend = df.groupby(['year', 'sex'])['suicides/100k pop'].mean().reset_index()

## Because the Data of 2016 is not complete (Sample Size Bias)
gap_trend_clean = gap_trend[gap_trend['year'] < 2016]

px.line(gap_trend_clean, x='year', y='suicides/100k pop', color='sex', 
        title='Gender Gap Trend (1985 - 2015) - Excluding Incomplete 2016 Data')
```

- Analysis Conclusion: > The gender gap widened significantly between 1985 and 1995, as male suicide rates surged while female rates remained relatively stable. Although there was a gradual decline for both genders after 1995, the gap remains substantial, with male rates consistently staying 3 to 4 times higher than female rates. A sharp increase is also observed in 2016, which requires further investigation to see if it marks a new rising trend.



```python
# Generational Trends: Which generation shows the highest suicide risk?
generation_trend = df.groupby(['generation', 'sex'])['suicides/100k pop'].mean().reset_index()
px.bar(generation_trend, x='generation', y='suicides/100k pop', color='sex', title='Suicide Rates by Generation')
```

- Finding: Suicide risk is strongly associated with older generations. The G.I. Generation and Silent Generation exhibit the highest suicide rates globally. There is a clear downward trend as we move toward younger generations (Millennials and Gen Z). This reinforces our previous finding that suicide rates increase with age.

```python
# Global Trajectory: How has the total number of suicides evolved from 1985 to 2016? Is the global trend increasing or decreasing?
global_trend = df[df['year'] < 2016].groupby('year')['suicides_no'].sum().reset_index()

px.line(global_trend, x='year', y='suicides_no', title='Global Trajectory of Suicides (Total Numbers: 1985 - 2015)',markers=True)
```

- 1985-1999: A significant upward trend, with global suicide counts nearly doubling.

Peak Period: The numbers reached their maximum between 1999 and 2003.

Current Trend: Since 2003, there has been a notable decrease in the total number of suicides globally, returning to levels near those of the early 1990s. This overall trajectory suggests that while the issue peaked at the turn of the millennium, global efforts or social changes have successfully started to reduce the numbers in the last decade.

```python
# Which 10 countries have the highest total suicides_no, and how does this list change when looking at suicides/100k pop?
top_10_countries = df.groupby('country')['suicides_no'].sum().sort_values(ascending=False).head(10).reset_index()
top_10_countries_by_rate = df.groupby('country')['suicides/100k pop'].mean().sort_values(ascending=False).head(10).reset_index()
print(top_10_countries)
print('-'*100)
print(top_10_countries_by_rate)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=2, subplot_titles=("Top 10 by Total Suicides", "Top 10 by Rate (per 100k)"),horizontal_spacing=0.15)

fig.add_trace(go.Bar(x=top_10_countries['country'], y=top_10_countries['suicides_no'], name="Total"),row=1, col=1)

fig.add_trace(go.Bar(x=top_10_countries_by_rate['country'], y=top_10_countries_by_rate['suicides/100k pop'], name="Rate"),row=1, col=2)

fig.update_layout(height=500, width=1200, title_text="Geographic Comparison: Totals vs. Rates", showlegend=False)
fig.show()
```

- Geographic Insight:
While the Russian Federation, USA, and Japan record the highest total number of suicides, the picture changes drastically when looking at the suicide rate per 100k population. Lithuania emerges as the highest-risk country globally, followed by Sri Lanka. Interestingly, Russia is the only country that appears in the top 3 of both lists, indicating it faces both a high total burden and a high individual risk. This confirms that total counts are often biased by population size, making rates a more accurate metric for comparing national risk levels.

```python
#Do countries with similar cultures or geographic locations (e.g., Nordic countries or Middle Eastern countries) share similar suicide rate patterns?
middle_east_list = ['Bahrain', 'Kuwait', 'Oman', 'Qatar', 'United Arab Emirates', 'Turkey']
# feature Engineering 
df['Category'] = 'Rest of the World' 
df.loc[df['country'].isin(middle_east_list), 'Category'] = 'Middle East' 
comparison = df.groupby(['year', 'Category'])['suicides/100k pop'].mean().reset_index()

px.line(comparison, x='year', y='suicides/100k pop', color='Category',
        title='The Cultural Impact: Middle East vs. Global Average',
        line_shape='spline', 
        markers=True)
```

- Key Finding:
There is a profound and consistent disparity between the Middle East and the Rest of the World. The Middle East maintains a significantly lower suicide rate (consistently below 3 per 100k), showcasing a remarkable level of cultural resilience. While the global average experienced sharp increases in the 1990s due to socio-economic crises, the Middle Eastern trend remained relatively flat and stable. This suggests that strong social, religious, and family-oriented structures in the region act as a powerful deterrent against suicide, regardless of global external shifts.

```python
#Wealth Correlation: Does a higher gdp_per_capita (individual wealth) correlate with lower suicide rates, or is the opposite true?

gdp_analysis = df.groupby('country').agg({'suicides/100k pop': 'mean','gdp_per_capita ($)': 'mean'}).reset_index()

fig = px.scatter(gdp_analysis,x='gdp_per_capita ($)', y='suicides/100k pop',hover_name='country',trendline="ols", title='Correlation: Individual Wealth vs. Suicide Rates')

fig.show()
```

- Critical Finding:
The correlation between GDP per capita and suicide rates is remarkably weak/neutral. The trend line is almost flat, indicating that individual wealth is NOT a primary predictor of suicide risk. We see high-income countries and low-income countries sharing similar rates. This suggests that socio-economic status alone cannot explain the phenomenon; instead, cultural, psychological, and social support factors likely play a more dominant role.

```python
korea_data = df[df['country'] == 'Republic of Korea'].groupby('year').agg({
    'suicides/100k pop': 'mean',
    'gdp_per_capita ($)': 'mean'
}).reset_index()

fig = make_subplots(rows=2, cols=1, 
                    subplot_titles=("Growth of Wealth (GDP per Capita)", "Growth of Suicide Rates"),
                    vertical_spacing=0.15)

fig.add_trace(
    go.Scatter(x=korea_data['year'], y=korea_data['gdp_per_capita ($)'], name="Wealth", line=dict(color='green')),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=korea_data['year'], y=korea_data['suicides/100k pop'], name="Suicide Rate", line=dict(color='red')),
    row=2, col=1
)

fig.update_layout(height=700, title_text="The South Korean Paradox: Wealth vs. Mental Health", showlegend=False)
fig.show()
```

- Case Study Conclusion (South Korea):
South Korea presents a unique paradox. While the GDP per capita grew exponentially (from $3k to $30k), the suicide rate followed the same upward trajectory, peaking at over 40 per 100k. This directly answers the "Economic Transition" question: Rapid wealth accumulation does not always correlate with better mental health. In fact, it suggests that intense economic competition and social shifts during rapid development can act as significant stressors, leading to higher suicide rates.
